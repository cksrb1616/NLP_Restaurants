```
import requests
import json

APIKEY = ""
rest_list = []

# lat should be between 52.28, 52.40
# lng should be between 5.01,  4.75
list_loc = []
start_lat = 52.28
start_lng = 4.75
target_lat = start_lat
target_lng = start_lng
num_lat = round((52.40-52.28)/0.01 + 1)
num_lng = round((5.01-4.75)/0.01 + 1)
total_map = [[()] *num_lat for _ in range(num_lng)]

for i in range(num_lng):
  for j in range(num_lat):
    total_map[i][j] = (round(4.75+0.01*i,2), round(52.28+0.01*j,2))

def findPlaces(loc=("52.28","4.75"), radius=2000, pagetoken = None):
   lat, lng = loc
   type = "restaurant"
   url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json?location={lat},{lng}&radius={radius}&type={type}&key={APIKEY}{pagetoken}".format(lat = lat, lng = lng, radius = radius, type = type,APIKEY = APIKEY, pagetoken = "&pagetoken="+pagetoken if pagetoken else "")
   print(url)
   response = requests.get(url)
   res = json.loads(response.text)
   print("here results ---->>> ", len(res["results"]))

   l = ['name','rating','types']
   for i in range(len(res['results'])):
     a = [value for (key,value) in res['results'][i].items() if key in l]
     rest_list.append(a)
   pagetoken = res.get("next_page_token",None)
   print("here -->> ", pagetoken)
   return pagetoken

pagetoken = None


for i in range(num_lng):
  for j in range(num_lat):
    pagetoken = findPlaces(loc = (total_map[i][j][1],total_map[i][j][0]),pagetoken=pagetoken)
```
